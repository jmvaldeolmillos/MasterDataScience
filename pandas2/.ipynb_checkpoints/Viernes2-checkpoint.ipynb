```python
# EvoluciÃ³n de vuelos de un dÃ­a para otro. Uso de Merge para hacer un join.
```

```python
df1 = DataFrame(np.random.randint(0,10,3*5).reshape(3,5))
df1.columns = ['c1','c2','c3','c4','c5']
df1
```

```python
df1['key1'] = ['a','a','b']
```

```python
df1
```

```python
df2 = DataFrame(np.random.randint(0,10,3*5).reshape(3,5))
df2.columns = ['c1','c2','c3','c4','c5']
df2['key1'] = ['b','a','b']
```

```python
df2
```

```python
# Join entre los dos dataframe
pd.merge(df1,df2, on='key1')
```

```python
pd.merge(df1,df2, on='key1', how='inner')
```

```python
pd.merge(df1,df2, on='key1', how='left')
```

```python
pd.merge(df1,df2, on='key1', how='right')
```

```python
pd.merge(df1,df2, on='key1', how='outer')
```

```python
# Podemos seleccionar columnas en caso de que no haya ninguna comÃºn.
df2['key2'] = df2['key1']
```

```python
pd.merge(df1, df2, left_on='key1', right_on='key2')
```

```python
pd.merge(df1, df2, left_on='key1', right_on='key2', suffixes=('izda','dcha'))
```

```python
# AÃ±adir una fila y key2 tenga 
df2
```

```python
df2.append({'c1':2,'c2':3,'c3':5,'c4':6,'c5':2,'key1':'d','key2':'c'},ignore_index= True)
```

```python
df1['key2'] = 'e'
```

```python
df1
```

```python
# apilar dataframes
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.merge(df1, df2, left_on='key1', right_on='key2', how='inner')
```

```python
pd.merge(df1, df2, left_on='key1', right_on='key2', how='outer')
```

```python
pd.merge(df22jan, df23jan, on='Carrier')
```

```python
evodf = pd.merge(df22jan, df23jan, on='Carrier')
```

```python
evodf.columns = ['Carrier','22Jan','23Jan']
```

```python
evodf
```

```python
# Evol. mes actual - mes anterior / mes anterior
evodf['Evol'] = ((evodf['23Jan'] - evodf['22Jan'])/evodf['22Jan']) * 100
evodf
```

```python
def dec2(n):
    return round(n,2)
```

```python
# esto es para aplicar apply porque se hace directo.
# evodf['Evol'] = ((evodf['23Jan'] - evodf['22Jan'])/evodf['22Jan']) * 100.round(2)
evodf["Evolution"] = evodf.Evol.apply(dec2)
```

```python
evodf
```

```python
evodf.sort("Evolution", ascending = False)
```

```python
def evo(s):
    return (s['23Jan']- s['22Jan'])/s['22Jan']
```

```python
fila = evodf.ix[9]
```

```python
fila
```

```python
evo(fila)
```

```python
# Ejecutamos a todas las filas del dataframe
evodf.apply(evo,axis = 1)
```

```python
evodf
```

```python
evodf["Evolution2"] = evodf.apply(evo, axis=1)
```

```python
evodf
```

```python
# Plotear evoluciÃ³n 3
loc = range(len(evodf))
plt.bar(loc, evodf.Evolution)
plt.xticks(loc, evodf.Carrier)
plt.show()
```

```python
# Plotear evoluciÃ³n 3 ordenado.
# Importante el reset_index, sino no mueve las etiquetas de abajo!!!!!!!!!!
evodfs = evodf.sort("Evolution", ascending = False).reset_index()
loc = range(len(evodfs))
plt.bar(loc, evodfs.Evolution)
plt.xticks(loc, evodfs.Carrier)
plt.show()
```
