```python
# plt.plot(x, y, label="")
x = [1,2,3,4]
y = [3,5,6.6,3.1415]
```

```python
plt.plot(x,y,label='Lineal')
```

```python
# para evitar el error de arriba se usa, que lo nuestra dentro del notebook:
%matplotlib inline
```

```python
plt.plot(x,y,label='Linal')
```

```python
x2 = [1,2,3,4]
y2 = [1,2,3,4]
plt.plot(x,y,label='Linea 1')
# el x2 no hacía falta, puede valer x
plt.plot(x2,y2,label='Linea 2')
```

```python
# añadimos opciones antes de plotear.
plt.title("First Graph")
plt.xlabel('Indice')
plt.ylabel('Valores')

plt.plot(x,y,label='Linea 1')
plt.plot(x,y2,label='Linea 2')

# Despues de la línea o líneas se añade legend
# frameon = recuadra la legend.

plt.legend(loc='best', frameon=False)
```

```python
# Scatterplot
plt.title("First Graph")
plt.xlabel('Indice')
plt.ylabel('Valores')

plt.plot(x,y,"r--",label='Linea 1')
plt.plot(x,y2,label='Linea 2')

plt.legend(loc='best', frameon=False)
```
