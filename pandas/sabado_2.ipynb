```python
# debería ser un valor de tiempo, pero está como número.
```

```python
# Vamos a crear una columna hora de depTime
```

```python
def hora(number):
    return int(number/100)
```

```python
# df2['DepTime'].map(hora) Da error porque hay NaN
df2['DepTime'].fillna(0.0).map(hora)
```

```python
df2['DepHour'] = df2['DepTime'].fillna(0.0).map(hora)
```

```python
# Uso de Groupby
# A parte de líneas problemáticas podemos sacar aviones problemáticos.
# Agrupamos en base a tail number y sacamos luego el delay medio de cada grupo.

grouped_by_plane = df2.groupby('TailNum')
```

```python
grouped_by_plane['DepDelay']
```

```python
delays_by_plane = grouped_by_plane['DepDelay'].mean()
```

```python
mean_delays = delays_by_plane.copy()
```

```python
mean_delays.sort(ascending=False)
```

```python
mean_delays
```

```python
# Para contextualizar la media, usaremos el número de vuelos, pues hay valores muy grandes arriba.
means_counts = grouped_by_plane['DepDelay'].agg(['mean','count'])
```

```python
means_counts.head()
```

```python
# volvemos a ordenar.
```

```python
means_counts.sort('mean', ascending=False)
```

```python
sorted_means_count = means_counts.sort('mean', ascending=False)
```

```python
# Quitamos los de pocos vuelos porque desvirtúan, no son representativos.

sorted_means_count[sorted_means_count['count']>10]
```
